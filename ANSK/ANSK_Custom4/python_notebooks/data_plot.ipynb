```python
import sys
sys.path.append('../python/')

#matplotlib for plotting
import matplotlib as mpl
from matplotlib import pyplot as plt
plt.style.use('../mplstyles/stylelib/standard.mplstyle')

#other computational libraries
import numpy as np
from scipy import special
import math

#libraries for reading csv
import csv
import pandas as pd
```

```python
#Get and arrange data
sourcedata = pd.read_csv("../data/source_Na22/00-05-06_t250_i50-20-200_h50_g100-5.csv",header=None,usecols=[0])
#backgrounddata = pd.read_csv("../data/background/approx00-05-45_t250_i100-80-200-h50_g100-5.csv",header=None,usecols=[0])

#change to numpy
#sourcedata = sourcedata.values - 200*8192
#backgrounddata = backgrounddata.values - 200*8192 -50000
#print(np.max(backgrounddata))


#sourcedata = np.subtract(sourcedata,200*8192) 	#subtract the baseline (8192) times the gate value.
#backgrounddata = np.subtract(backgrounddata,200*8192)
#bins = np.arange(min(backgrounddata.min(),sourcedata.min()),max(backgrounddata.max(),sourcedata.max()))
```

```python
#translate to histogram

Emax = 700
Emin = 400

Ebins = np.linspace(Emin,Emax,1000)
xcq = (Ebins[:-1] + Ebins[1:]) / 2


#En,Enx = np.histogram(backgrounddata,bins=Ebins)
Ens,Enxs = np.histogram(sourcedata,bins=Ebins)
```

### Making a Fixed Plot Using Imported Values

```python
##set up a 1d plot
fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),sharex=True)
ax1 = axes


estring = r'${}\mathrm{{keV}}< E_r \leq {}\mathrm{{keV}}$'.format(Emin,Emax)
#ax1.step(xcq,En, where='mid',color='b', linestyle='-', \
#            label='Background', linewidth=2)
ax1.step(xcq,Ens, where='mid',color='orange', linestyle='-', \
            label='$^{22}$Na', linewidth=2)

ymin = 0.0
ymax = 300

blue = '#118DFA'
#ax1.fill_between(xcq,np.zeros(np.shape(xcq)),En,step='mid',facecolor='m',alpha=0.4, \
#                label='')
ax1.fill_between(xcq,np.zeros(np.shape(xcq)),Ens,step='mid',facecolor='orange',alpha=0.4, \
                 label='')

#ax1.axvline(Q,color='k',linestyle='--',linewidth=2,label='Q-value')


ax1.set_yscale('linear')
#ax1.set_yscale('log')
ax1.set_xlim(Emin, Emax) 
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'deposited energy [bins]')
ax1.set_ylabel('counts')
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')
ax1.legend(loc=1,prop={'size':22})
#ax1.legend(bbox_to_anchor=(1.04,1),borderaxespad=0,prop={'size':22})

for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
plt.savefig('figures/ANSK_Custom4_22Nai50.png')
plt.show()
```
