# Detector Rates
The purpose of this notebook is to compare a scan over the same window (lower threshold 5000) for different subsets of the detector array.

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import data
detall = pd.read_csv('data/NaIscanwindow_20220426_4_out.csv')
det1_4 = pd.read_csv('data/NaIscanwindow_20220503_Det1-4_out.csv')
det5_8 = pd.read_csv('data/NaIscanwindow_20220503_Det5-8_out.csv')
det9_12 = pd.read_csv('data/NaIscanwindow_20220503_Det9-12_out.csv')
det13_16 = pd.read_csv('data/NaIscanwindow_20220503_Det13-16_out.csv')
det21_23 = pd.read_csv('data/NaIscanwindow_20220503_Det21-23_out.csv')

det5 = pd.read_csv('data/NaIscanwindow_20220503_Det5_out.csv')

x = detall.iloc[:,0]
detsum_y = det1_4.iloc[:,1] + det5_8.iloc[:,1] + det13_16.iloc[:,1] + det21_23.iloc[:,1]
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,detall.iloc[:,1],label="All Detectors",alpha=0.3,color='grey')
plt.scatter(x,det1_4.iloc[:,1],label="Detectors 1-4",alpha=0.6,marker='s')
plt.scatter(x,det5_8.iloc[:,1],label="Detectors 5-8",alpha=0.6,marker='s')
plt.scatter(x,det9_12.iloc[:,1],label="Detectors 9-12",alpha=0.6,marker='s')
plt.scatter(x,det13_16.iloc[:,1],label="Detectors 13-16",alpha=0.6,marker='s')
plt.scatter(x,det21_23.iloc[:,1],label="Detectors 21-23",alpha=0.6,marker='s',edgecolors='black')
plt.scatter(x,det5.iloc[:,1],label="Detector 5",marker='v')
plt.scatter(x,detsum_y,label="Sum of Subsets",color='black')

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,det5.iloc[:,1],label="Detector 5",marker='v')

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold: Detector 5 Only",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,det21_23.iloc[:,1],label="Detectors 21-23",alpha=0.6,marker='s',edgecolors='black')

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold: Detectors 21-23 Only",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.show()
```
