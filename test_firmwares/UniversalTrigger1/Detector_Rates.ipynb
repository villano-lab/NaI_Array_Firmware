# Detector Rates
The purpose of this notebook is to compare a scan over the same window (lower threshold 5000) for different subsets of the detector array.

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import data
detall = pd.read_csv('data/NaIscanwindow_20220426_4_out.csv')

det1_4 = pd.read_csv('data/NaIscanwindow_20220503_Det1-4_out.csv').iloc[:,1]
det5_8 = pd.read_csv('data/NaIscanwindow_20220503_Det5-8_out.csv').iloc[:,1]
det9_12 = pd.read_csv('data/NaIscanwindow_20220503_Det9-12_out.csv').iloc[:,1]
det13_16 = pd.read_csv('data/NaIscanwindow_20220503_Det13-16_out.csv').iloc[:,1]
det17_20 = pd.read_csv('data/NaIscanwindow_20220503_Det17-20_out.csv').iloc[:,1]
det21_23 = pd.read_csv('data/NaIscanwindow_20220503_Det21-23_out.csv').iloc[:,1]

det20_22 = pd.read_csv('data/NaIscanwindow_20220504_Det20-22_out.csv').iloc[:,1]

det5 = pd.read_csv('data/NaIscanwindow_20220503_Det5_out.csv').iloc[:,1]
det21 = pd.read_csv('data/NaIscanwindow_20220504_Det21_out.csv').iloc[:,1]
det22 = pd.read_csv('data/NaIscanwindow_20220504_Det22_out.csv').iloc[:,1]
det23 = pd.read_csv('data/NaIscanwindow_20220504_Det23_out.csv').iloc[:,1]

x = detall.iloc[:,0] #All data is taken over the same set of x values
detsum_y = det1_4 + det5_8 + det9_12 + det13_16 + det17_20 + det21_23
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,detall.iloc[:,1],label="All Detectors",alpha=0.3,color='grey')
plt.scatter(x,det1_4,label="Detectors 1-4",alpha=0.6,marker='s')
plt.scatter(x,det5_8,label="Detectors 5-8",alpha=0.6,marker='s')
plt.scatter(x,det9_12,label="Detectors 9-12",alpha=0.6,marker='s')
plt.scatter(x,det13_16,label="Detectors 13-16",alpha=0.6,marker='s')
plt.scatter(x,det17_20,label="Detectors 17-20",alpha=0.6,marker='s')
plt.scatter(x,det21_23,label="Detectors 21-23",alpha=0.6,marker='s',edgecolors='black',color="red")
plt.scatter(x,det5,label="Detector 5",marker='v',color="purple")
plt.scatter(x,detsum_y,label="Sum of Subsets",color='black')

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(fontsize=18,loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,det5,label="Detector 5",marker='v',color="purple",edgecolors="black")
plt.scatter(x,det21,label="Detector 21",marker='v',alpha=0.6)
plt.scatter(x,det22,label="Detector 22",marker='v',alpha=0.6)
plt.scatter(x,det23,label="Detector 23",marker='v',alpha=0.6)

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold: Single Detectors",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,det21_23,label="Detectors 21-23",alpha=0.6,marker='s',edgecolors='black',color="red")
plt.scatter(x,det20_22,label="Detectors 20-22",alpha=0.6,marker='s',edgecolors='black')

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold: Groups of Three",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(16,9))

#plt.scatter(x,3*det5,label="3*Detector 5",marker='v',color="purple")
plt.scatter(x,det21_23,label="Detectors 21-23",alpha=0.6,marker='s',edgecolors='black',color="red")
plt.scatter(x,det21+det22+det23,label="Sum of Detectors 21, 22, and 23",color="black")

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold: Individual vs. Group",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(loc='upper left')
plt.show()
```
