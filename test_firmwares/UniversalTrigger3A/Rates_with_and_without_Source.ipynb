```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline as interp
```

```python
#Import data
data_bg = pd.read_csv('data/no_PuBe_scan.csv')
data_floor = pd.read_csv('data/PuBe_floor_scan.csv')
data_raised = pd.read_csv('data/PuBe_raised_scan.csv')
data_UT2 = pd.read_csv('../UniversalTrigger2/data/scanwindow_06052022.csv') #for comparison

threshold_3A = data_bg.iloc[:,0] #used same set of thresholds for all of them.
threshold_2 = data_UT2.iloc[:,0]

rate_bg = data_bg.iloc[:,1]
rate_floor = data_floor.iloc[:,1]
rate_raised = data_raised.iloc[:,1]
rate_2 = data_UT2.iloc[:,1]

del data_bg, data_floor, data_raised, data_UT2
```

```python
plt.figure(figsize=(16,9))

plt.scatter(threshold_3A,rate_bg,label='Background')
plt.scatter(threshold_3A,rate_floor,label='Floor')
plt.scatter(threshold_3A,rate_raised,label='Raised')
plt.scatter(threshold_2,rate_2,label='UT2, L.Thresh 4499',)

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels

plt.legend()
plt.show()
```

*Note: Lower threshold was 4337 unless stated otherwise.*
