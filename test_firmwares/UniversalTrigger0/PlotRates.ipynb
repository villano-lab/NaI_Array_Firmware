```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import norm

data = pd.read_csv("data/NaIscanwindow_20220426_2_out.csv") #Change to match local file name/location
data2 = pd.read_csv("data/NaIscanwindow_20220426_3_out.csv") #Change to match local file name/location
```

```python
plt.figure(figsize=(16,9))

plt.scatter(data.iloc[:,0],data.iloc[:,1],s=6,label="4192 window trigger")
plt.scatter(data2.iloc[:,0],data2.iloc[:,1],s=6,label="5000 window trigger")
#plt.plot(data.iloc[:,0],data.iloc[:,1],label="interpolation",zorder=0,color="#AAAAAA")
plt.axvline(8192,color="green",label="8192")

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(4000)
plt.ylim(0)

plt.legend(fontsize=18)
plt.show()
```
