# Detector Rates
The purpose of this notebook is to compare a scan over the same window (lower threshold 5000) for different subsets of the detector array.

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import data
original = pd.read_csv('../UniversalTrigger1/data/NaIscanwindow_20220426_4_out.csv')
original100 = pd.read_csv('../UniversalTrigger1/data/NaIscanwindow_20220513_all_out.csv')
new = pd.read_csv('data/NaIscanwindow_20220517_out.csv')
mimic = pd.read_csv('data/NaIscanwindow_20220517_2_out.csv')
rerun = pd.read_csv('../UniversalTrigger1/data/NaIscanwindow_20220518_UT1_out.csv')
lower0 = pd.read_csv('data/NaIscanwindow_20220526_out.csv')

x = original.iloc[:,0] #All data is taken over the same set of x values
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,original.iloc[:,1],label="Upper Gate 10",alpha=0.6,marker='s')
plt.scatter(x,original100.iloc[:,1],label="Uppger Gate 100 (5/13)",alpha=0.6,marker='s')
plt.scatter(x,rerun.iloc[:,1],label="Upper Gate 100 (5/18)",alpha=0.6,marker='s')
plt.scatter(x,new.iloc[:,1],label="Upper Gate 100, Lower Gate 50",alpha=0.6)
plt.scatter(x,mimic.iloc[:,1],label="Upper Gate 100, Lower Gate 1, Bugfix 1",alpha=0.6)
plt.scatter(x,lower0.iloc[:,1],label="Upper Gate 100, Lower Gate 0",alpha=0.8)

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Upper Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.xlim(5000)
plt.ylim(0)

plt.legend(fontsize=18,loc='upper left')
plt.show()
```
