# Detector Rates
The purpose of this notebook is to plot results of various scans.

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline as interp
```

```python
#Import data
data = pd.read_csv ("data/scanrate_06022022_3.csv")
threshold = data.iloc[:,0]
rate = data.iloc[:,1]

rate = rate[threshold > 220]
threshold = threshold[threshold > 220]
```

```python
plt.figure(figsize=(16,9))

plt.scatter(threshold,rate)


plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate on Background vs Lower Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
plt.yscale('log')
plt.xlim(150)
#plt.ylim(0)

#plt.legend(fontsize=18,loc='upper left')
plt.show()
```

```python
diffs = np.zeros(len(threshold))
for i,n in enumerate(rate):
    try:
        diffs[i] = rate[i-1] - n
    except KeyError:
        diffs[i] = np.nan
```

```python
plt.figure(figsize=(16,9))

plt.scatter(threshold,diffs)


plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Difference in Trigger Rate on Background vs Lower Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
#plt.yscale('log')
#plt.xlim(5000)
#plt.ylim(0)
plt.axvline(700)
plt.axvline(220)
plt.axvline(880)
plt.axvline(1940)
plt.axvline(2650)
point = [880,511]
point2= [1940,1274]
#plt.legend(fontsize=18,loc='upper left')
plt.show()
```

```python
plt.plot(point[0],point[1],"o")
plt.plot(point2[0],point2[1],"o")

m = (point2[1] - point[1])/(point2[0]-point[0])
b = point2[1] - m*point2[0]
def y(x):
    return m*x+b
np.arange(500,2000)
x = np.arange(500,2000)
plt.plot(x,y(x))
print ((3000 - b)/(m))
y(2650)
```

```python
interpolated = interp(threshold,rate)
derivspline = interpolated.derivative()
derivative = - derivspline(threshold)
#print(len(derivative),len(threshold))
```

```python
plt.figure(figsize=(16,9))

plt.scatter(threshold,derivative)


plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Derivative of Trigger Rate on Background vs Lower Threshold",size=30)
plt.rc('xtick', labelsize=16) #fontsize of the x tick labels
plt.rc('ytick', labelsize=16) #fontsize of the y tick labels
#plt.yscale('log')
#plt.xlim(5000)
#plt.ylim(0)

#plt.legend(fontsize=18,loc='upper left')
plt.show()
```
