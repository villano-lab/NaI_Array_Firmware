```python
import pandas as pds
import numpy as np

data = pds.read_csv(r"C:\Users\hellmala\Desktop\stuff\FIFOtest6td.txt" , delimiter="\t")
#print(data)
#dataarr=np.asarray(data)
datalist = data['00000000'].tolist() 
print(datalist[4])

#remove the garbage at the end
datalist = [s[:-4] for s in datalist]
print(datalist[4])
```

```python
#times of each gamma event 
timetag=[]
for i in range(1, len(datalist), 3):
    timetag.append(datalist[i])

#for i in range(1, len(datalist), 3):
 #   if not pds.isna(datalist[i]) and all(c in '0123456789ABCDEFabcdef' for c in datalist[i][0]):
  #      timetag.append(int(datalist[i][0], 16))
   # else:
        # Add a placeholder for non-valid hexadecimal strings
    #    timetag.append('xxxxx')
    
#print('hi1')
print(timetag)

# Convert each hexadecimal timetag to an integer
timetag_int=[]
for hex_str in timetag:
    if all(c in '0123456789ABCDEFabcdef' for c in hex_str):
        timetag_int.append(int(hex_str, 16))
    else:
        # for non hexidecimal strings
        timetag_int.append(None)
        
        
#print('hi2')
#print(timetag_int)

#gamma energies
energy=[]
for i in range(2, len(datalist),3):
    energy.append(datalist[i])
    
#energy hex to int    
E_int=[]
for hex_str in energy:
    if all(c in '0123456789ABCDEFabcdef' for c in hex_str):
        E_int.append(int(hex_str, 16))
    else:
        # for non hexidecimal strings
        E_int.append(None)
        


#get rid of those weird non hexidecimals stuff
E_int = [i for i in E_int if 'x' not in str(i)]

```

```python
import matplotlib.pyplot as plt

#energies histogram/make the spectra
plt.hist(E_int, bins=100, edgecolor='k', range=(10000000,50000000))
plt.xlabel('E')
plt.ylabel('counts')
plt.title('Spectrum')
plt.grid(True)

plt.show()
```
